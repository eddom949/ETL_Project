```python
csv_file = "../Resources/IRSIncomeByZIPCode.csv"
income_df = pd.read_csv(csv_file)
income_df.head()
```

```python
new_income_df = income_df[['STATE','ZIPCODE','Avg taxable income']]
new_income_df.head()
```

```python
new_income_df.drop(new_income_df.loc[new_income_df['ZIPCODE']==0].index, inplace=True)
new_income_df.head()
```

```python
connection = 'postgres:1difalatty@localhost:5432/zomato_db'
engine = create_engine(f'postgresql://{connection}')
```
